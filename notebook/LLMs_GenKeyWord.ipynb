```
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM, BitsAndBytesConfig

model_id = "Qwen/Qwen3-32B"
print(f"Downloading {model_id} to 4-bit")

bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_use_double_quant=True,
    bnb_4bit_compute_dtype=torch.bfloat16
)

tokenizer = AutoTokenizer.from_pretrained(model_id)

model = AutoModelForCausalLM.from_pretrained(
    model_id,
    quantization_config=bnb_config,
    device_map="auto",
    trust_remote_code=True
)

print(f"Downloaded Qwen 3 ({model.get_memory_footprint() / 1e9:.2f} GB VRAM)!")
```

```
import json
from tqdm import tqdm

# Danh sách 5 công ty
TARGET_COMPANIES = [
    {"ticker": "VIC", "name": "Tập đoàn Vingroup", "industry": "Bất động sản & Đa ngành"},
    {"ticker": "HPG", "name": "Tập đoàn Hòa Phát", "industry": "Thép & Vật liệu xây dựng"},
    {"ticker": "FPT", "name": "Công ty Cổ phần FPT", "industry": "Công nghệ thông tin"},
    {"ticker": "VNM", "name": "Vinamilk", "industry": "Thực phẩm & Đồ uống"},
    {"ticker": "MWG", "name": "Thế Giới Di Động", "industry": "Bán lẻ"}
]

def generate_keywords_qwen3(ticker, company_name, industry):
    # Prompt Template (Kiểm tra xem Qwen3 dùng Chat Template nào, thường là ChatML)
    # Nếu là Base model (không phải Instruct), ta phải dùng kỹ thuật prompting khác
    # Ở đây giả định bạn đang dùng bản có khả năng chat/instruct

    prompt = f"""<|im_start|>system
Bạn là chuyên gia phân tích dữ liệu tài chính cấp cao tại thị trường chứng khoán Việt Nam (HOSE/HNX).
Nhiệm vụ của bạn là tạo ra danh sách từ khóa tìm kiếm (Search Keywords) có độ chính xác cao (High Signal-to-Noise Ratio).
Tuyệt đối không bịa đặt tên người hoặc liệt kê các công ty không cạnh tranh trực tiếp.<|im_end|>
<|im_start|>user
Hãy sinh bộ từ khóa tìm kiếm tin tức lịch sử (2018-2024) cho công ty sau:
- Mã CK: {ticker}
- Tên: {company_name}
- Ngành kinh doanh: {industry}

Yêu cầu chi tiết cho 3 nhóm từ khóa (Output JSON Tiếng Việt):

1. **"keywords_direct" (Trực tiếp):**
   - Bao gồm: Tên công ty (viết tắt/đầy đủ), Mã CK, Tên Chủ tịch/CEO *thực tế* (chính xác 100%), Tên các dự án/sản phẩm trọng điểm (cả cũ và mới).
   - LƯU Ý: Không đưa tên viết tắt 3 chữ cái nếu nó dễ gây nhầm lẫn (Ví dụ: Tránh 'VAS', 'VFC' trừ khi chắc chắn). Không bịa tên lãnh đạo không nổi tiếng.

2. **"keywords_competitors" (Đối thủ & Đối tác):**
   - Chỉ liệt kê các đối thủ cạnh tranh **trực tiếp tại thị trường Việt Nam** trong các mảng kinh doanh cốt lõi.
   - KHÔNG liệt kê các tập đoàn công nghệ/tiêu dùng toàn cầu (Apple, Google, Microsoft, Coca-Cola, Pepsi) trừ khi họ cạnh tranh trực tiếp thị phần với {ticker} (Ví dụ: Nếu {ticker} bán điện thoại thì mới kể Apple, nếu bán nhà thì ĐỪNG kể Apple).
   - Nếu là tập đoàn đa ngành (như Vingroup), hãy liệt kê đối thủ cho cả mảng Bất động sản VÀ Xe điện.

3. **"keywords_macro" (Vĩ mô & Chính sách):**
   - Phải bao phủ **TẤT CẢ** các lĩnh vực kinh doanh chính của công ty.
   - Ví dụ: Nếu công ty làm cả Xây dựng và Sản xuất, phải có từ khóa về "Giá thép", "Xi măng" VÀ "Luật đầu tư công".
   - Bao gồm: Chính sách thuế, Luật sửa đổi, Giá hàng hóa đầu vào (Commodities), Tỷ giá, Lãi suất.

OUTPUT JSON FORMAT ONLY:
{{
    "ticker": "{ticker}",
    "keywords_direct": ["..."],
    "keywords_competitors": ["..."],
    "keywords_macro": ["..."]
}}
<|im_end|>
<|im_start|>assistant
"""

    inputs = tokenizer(prompt, return_tensors="pt").to("cuda")

    outputs = model.generate(
        **inputs,
        max_new_tokens=1024,
        temperature=0.6,
        do_sample=True,
        pad_token_id=tokenizer.eos_token_id
    )

    response = tokenizer.decode(outputs[0][inputs.input_ids.shape[1]:], skip_special_tokens=True)
    return response

import re

def clean_json(text):
    match = re.search(r'\{.*\}', text, re.DOTALL)
    return match.group(0) if match else text

print("bắt đầu sinh keyword")

for company in tqdm(TARGET_COMPANIES):
    try:
        raw_text = generate_keywords_qwen3(company['ticker'], company['name'], company['industry'])

        # Parse JSON
        clean_data = clean_json(raw_text)
        json_data = json.loads(clean_data)

        path = f"{output_folder}/{company['ticker']}_keywords.json"
        with open(path, 'w', encoding='utf-8') as f:
            json.dump(json_data, f, ensure_ascii=False, indent=4)

        print(f"Saved: {company['ticker']}")

    except Exception as e:
        print(f"Lỗi {company['ticker']}: {e}")
        print(raw_text)
```
